```python
# 모의고사
def solve(pattern, answers):
    N = len(answers)
    M = len(pattern)
    cnt = 0
    for i in range(N):
        if pattern[i % M] == answers[i]:
            cnt += 1
    return cnt

def solution(answers):
    answer = []
    pattern_1 = [1, 2, 3, 4, 5]
    pattern_2 = [2, 1, 2, 3, 2, 4, 2, 5]
    pattern_3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]
    tmp = []
    tmp.append(solve(pattern_1, answers))
    tmp.append(solve(pattern_2, answers))
    tmp.append(solve(pattern_3, answers))
    m = max(tmp)
    for i in range(3):
        if tmp[i] == m:
            answer.append(i+1)
    return answer
```

```python
# 소수찾기 DFS로 permutation 사용
import math
# 선택할 수, visited, 목표 길이, 현재 string
def DFS(numbers, visited, M, string):
    if len(string) == M:
        n = int(string)
        return [n]
    tmp = []
    for i in range(len(numbers)):
        if visited[i] == 0:
            visited[i] = 1
            tmp += DFS(numbers, visited, M, string+numbers[i])
            visited[i] = 0
    return tmp
# numbers 길이
def check(n):
    k = math.sqrt(n)
    if n < 2: 
        return False

    for i in range(2, int(k)+1):
        if n % i == 0:
            return False
    return True
        

def solution(numbers):
    answer = 0
    N = len(numbers)
    #소수
    # prime = cprime(N)
    # 가능한 조합 모두 만들기
    data = set([])
    #i: 선택 가능한 문자 개수
    for i in range(1, N+1):
        visited = [0] * N
        tmp = DFS(numbers, visited, i, "")
        for el in tmp:
            if check(el) and el not in data:
                answer += 1
                data.add(el)
    return answer
```

```python
from itertools import permutations

def check(n):
    if n < 2:
        return False
    
    for i in range(2, int(n ** 0.5)+1):
        if n % i == 0:
            return False
    return True

def solution(numbers):
    answer = 0
    # str => list로 변화
    numbers = list(numbers)
    data = set([])
    for i in range(1, len(numbers)+1):
        tmp = list(permutations(numbers, i))
        for el in tmp:
            el = int("".join(el))
            # 예외 처리
            if check(el) and el not in data:
                data.add(el)
                answer += 1
            
    return answer
```

```python
# 체육복
def solution(n, lost, reserve):
    answer = 0
    
    # 셋 차집합 스킬
    lost, reserve = set(lost) - set(reserve), set(reserve) - set(lost)
    
    for el in reserve:
        if el-1 in lost:
            lost.remove(el-1)
        elif el+1 in lost:
            lost.remove(el+1)
    
                
    return n-len(lost)
```

```python
#조이스틱
def solution(name):
    answer = 0
    N = len(name)
    tmp = "A"*N
    
    c = 0
    while tmp != name:
        a = ord(name[c])
        answer += min(91 - a, a - 65)
        tmp = tmp[:c] + name[c] + tmp[c+1:]
        
        if tmp == name:
            break
        
        # 커서 이동
        l, r = (c-1) % N, (c+1) % N
        if tmp[l] != name[l] and tmp[r] != name[r]:
            c = r
            answer+=1
        elif tmp[l] != name[l] and tmp[r] == name[r]:
            c = l
            answer+=1
        elif tmp[l] == name[l] and tmp[r] != name[r]:
            c = r
            answer+=1
        else:
            # 둘다 맞는 경우 => 안맞는 곳 중 가장 가까운 곳
            a,b, s, k = 0, 0, 0, 0
            for i in range(N):
                a = (c+i) % N
                if tmp[a] != name[a]:
                    s = i
                    break
            for i in range(N):
                b = (c-i) % N
                if tmp[b] != name[b]:
                    k = i
                    break
            if s <=k:
                c = a
                answer += s
            else:
                c = b
                answer += k
                
    return answer
```

```python
# 큰수만들기
# 큐 사용 = > 시간복잡도 nk 근데 이 문제는 n번에 끝내는게 가능한 분제
from collections import deque
def solution(number, k):
    number = deque(list(number))
    answer = ''
    # 보관함
    que = []
    for _ in range(k):
        n = len(number)
        for i in range(n-1):
            if int(number[0]) >= int(number[1]):
                que.append(number.popleft())
            else:
                number.popleft()
                while number:
                    que.append(number.popleft())
                number = deque(que)
                que = []
                break
    answer = "".join(number)
    return answer
```
