```python
# 모의고사
def solve(pattern, answers):
    N = len(answers)
    M = len(pattern)
    cnt = 0
    for i in range(N):
        if pattern[i % M] == answers[i]:
            cnt += 1
    return cnt

def solution(answers):
    answer = []
    pattern_1 = [1, 2, 3, 4, 5]
    pattern_2 = [2, 1, 2, 3, 2, 4, 2, 5]
    pattern_3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]
    tmp = []
    tmp.append(solve(pattern_1, answers))
    tmp.append(solve(pattern_2, answers))
    tmp.append(solve(pattern_3, answers))
    m = max(tmp)
    for i in range(3):
        if tmp[i] == m:
            answer.append(i+1)
    return answer
```

```python
# 소수찾기 DFS로 permutation 사용
import math
# 선택할 수, visited, 목표 길이, 현재 string
def DFS(numbers, visited, M, string):
    if len(string) == M:
        n = int(string)
        return [n]
    tmp = []
    for i in range(len(numbers)):
        if visited[i] == 0:
            visited[i] = 1
            tmp += DFS(numbers, visited, M, string+numbers[i])
            visited[i] = 0
    return tmp
# numbers 길이
def check(n):
    k = math.sqrt(n)
    if n < 2: 
        return False

    for i in range(2, int(k)+1):
        if n % i == 0:
            return False
    return True
        

def solution(numbers):
    answer = 0
    N = len(numbers)
    #소수
    # prime = cprime(N)
    # 가능한 조합 모두 만들기
    data = set([])
    #i: 선택 가능한 문자 개수
    for i in range(1, N+1):
        visited = [0] * N
        tmp = DFS(numbers, visited, i, "")
        for el in tmp:
            if check(el) and el not in data:
                answer += 1
                data.add(el)
    return answer
```

```python
from itertools import permutations

def check(n):
    if n < 2:
        return False
    
    for i in range(2, int(n ** 0.5)+1):
        if n % i == 0:
            return False
    return True

def solution(numbers):
    answer = 0
    # str => list로 변화
    numbers = list(numbers)
    data = set([])
    for i in range(1, len(numbers)+1):
        tmp = list(permutations(numbers, i))
        for el in tmp:
            el = int("".join(el))
            # 예외 처리
            if check(el) and el not in data:
                data.add(el)
                answer += 1
            
    return answer
```

```python
# 체육복
def solution(n, lost, reserve):
    answer = 0
    
    # 셋 차집합 스킬
    lost, reserve = set(lost) - set(reserve), set(reserve) - set(lost)
    
    for el in reserve:
        if el-1 in lost:
            lost.remove(el-1)
        elif el+1 in lost:
            lost.remove(el+1)
    
                
    return n-len(lost)
```

```python
#조이스틱
def solution(name):
    answer = 0
    N = len(name)
    tmp = "A"*N
    
    c = 0
    while tmp != name:
        a = ord(name[c])
        answer += min(91 - a, a - 65)
        tmp = tmp[:c] + name[c] + tmp[c+1:]
        
        if tmp == name:
            break
        
        # 커서 이동
        l, r = (c-1) % N, (c+1) % N
        if tmp[l] != name[l] and tmp[r] != name[r]:
            c = r
            answer+=1
        elif tmp[l] != name[l] and tmp[r] == name[r]:
            c = l
            answer+=1
        elif tmp[l] == name[l] and tmp[r] != name[r]:
            c = r
            answer+=1
        else:
            # 둘다 맞는 경우 => 안맞는 곳 중 가장 가까운 곳
            a,b, s, k = 0, 0, 0, 0
            for i in range(N):
                a = (c+i) % N
                if tmp[a] != name[a]:
                    s = i
                    break
            for i in range(N):
                b = (c-i) % N
                if tmp[b] != name[b]:
                    k = i
                    break
            if s <=k:
                c = a
                answer += s
            else:
                c = b
                answer += k
                
    return answer
```

```python
# 큰수만들기
# 큐 사용 = > 시간복잡도 nk 근데 이 문제는 n번에 끝내는게 가능한 분제
from collections import deque
def solution(number, k):
    number = deque(list(number))
    answer = ''
    # 보관함
    que = []
    for _ in range(k):
        n = len(number)
        for i in range(n-1):
            if int(number[0]) >= int(number[1]):
                que.append(number.popleft())
            else:
                number.popleft()
                while number:
                    que.append(number.popleft())
                number = deque(que)
                que = []
                break
    answer = "".join(number)
    return answer
```

동적프로그래밍(DP)

```python
# N으로 표현 // 얼얼한 문제... dp 방향성이 이상하다 알았으면 바꿀방법을 생각하자..
# set(집합과)dp의 조합
# 사칙연산의 개념??
def solution(N, number):
    answer = -1
    dp = [set([]) for i in range(10)]
    dp[1].add(N)
    if number == N:
        return 1
    
    for i in range(2, 10):
        dp[i].add(N * int("1" * i))
        for j in range(1, i):
            k = i - j
            # dp[j]와 dp[k] 사칙연산
            # 이렇게 두번 (1, 2) (2, 1) 하는 이유는 +, *는 바뀌어도 연산이 같지만 -//는 바뀌면 값이 달라짐
            for el in dp[j]:
                for el2 in dp[k]:
                    dp[i].add(el + el2)
                    dp[i].add(el - el2)
                    dp[i].add(el * el2)
                    if el2 != 0:
                        dp[i].add(el // el2)
        if number in dp[i]:
            return i
        
    return answer
```

```python
# 정수삼각형
def solution(triangle):
    answer = 0
    N = len(triangle)
    # 슬라이딩 윈도우
    dp = [0] * (N+1)
    dp[1] = triangle[0][0]
    for i in range(2, N+1):
        for j in range(i, 0, -1):
            dp[j] = triangle[i-1][j-1] + max(dp[j], dp[j-1])

    answer = max(dp)
    
    return answer
```

```python
# 등굣길
def solution(m, n, puddles):
    # 1, 1시작
    answer = 0
    # 웅덩이 0 아님 1
    dp = [[1] * (m+1) for _ in range(n+1)]
    for i in range(m+1):
        dp[0][i] = 0
    for i in range(n+1):
        dp[i][0] = 0
    
    for i,j in puddles:
        dp[j][i] = 0
    
    for i in range(1, n+1):
        for j in range(1, m+1):
            if i == 1 and j == 1:
                continue
            if dp[i][j] != 0:
                dp[i][j] = (dp[i-1][j] + dp[i][j-1]) % 1000000007
    answer = dp[n][m]
    return answer
```

```python
# 도둑질
def solution(money):
    answer = 0
    N = len(money)
    # dp1: 첫번쨰 집은 털고 마지막 집은 안터는거
    # dp2: 마지막 집은 털고 첫번째 집은 안터는거
    # 원형이기 때문에 나오는 구조
    dp1 = [0] * (N+1)
    dp2 = [0] * (N+1)
    
    dp1[1] = money[0]
    
    #dp1 마지막 집은 안턴다
    for i in range(2, N):
        dp1[i] = max(dp1[i-1], dp1[i-2] + money[i-1])
    
    for i in range(2, N+1):
        dp2[i] = max(dp2[i-1], dp2[i-2] + money[i-1])
        
    answer = max(dp1[-2], dp2[-1])
    return answer
```

```python
# 타겟넘버
# data, 당해야할 idx, 현재 val
def DFS(data, idx, val, target):
    if idx == len(data):
        if val == target:
            return 1
        return 0
    tmp = 0
    tmp += DFS(data, idx+1, val + data[idx], target)
    tmp += DFS(data, idx+1, val - data[idx], target)
    return tmp

def solution(numbers, target): 
    answer = 0
    answer = DFS(numbers, 0, 0, target)
    return answer
```
